# A simple line chart

```python
L1=[1,2,3,4]
L2=[10,25,32,58]

plt.plot(L1,L2)
plt.show()

#When you run this code, a window should pop up displaying the line plot of L1 and L2. 
#The x-axis of the plot corresponds to the values in L1, and the y-axis corresponds to the values in L2. 
#The plot will show a line connecting the points (1, 10), (2, 25), (3, 32), and (4, 58).
```

# Subplots in the same figure

# First method

```python
x=np.arange(10)
y1=x+5
y2=x**2
plt.figure(figsize=(10,5))

plt.subplot(1,2,2)             #one row and 2 column  2 nd graph
plt.plot(x,y2,"g")

plt.subplot(1,2,1)              #one row and 2 column  1 st graph
plt.plot(x,y1,"r")

plt.show()
```

```python
x=np.arange(10)
y1=x+5
y2=x**2
plt.figure(figsize=(10,5))

plt.subplot(2,1,1)             #one row and 2 column 1st graph
plt.plot(x,y2,"g")

plt.subplot(2,1,2)             # two column 1 row 2nd graph
plt.plot(x,y1,"r")

plt.show()
```

```python
x=np.arange(10)
y1=x+5
y2=x**2
y3=-x
y4=np.sqrt(x)
plt.figure(figsize=(15,10))


plt.subplot(2,2,1)              # 2 rows 2 columns
plt.plot(x,y1,"g")


plt.subplot(2,2,2)
plt.plot(x,y2,"r")


plt.subplot(2,2,3)
plt.plot(x,y3,"b")


plt.subplot(2,2,4)
plt.plot(x,y4,"orange")

plt.show()
```

# Second method

```python
x=np.arange(10)
y1=-(x**2)
y2=x**3
y3=np.exp(x)
y4=np.sqrt(x)

fig,sfigs=plt.subplots(nrows=2,ncols=2,figsize=(10,7))

sfigs[0,0].plot(x,y1,"r")
sfigs[0,0].set_title("first graph")

sfigs[0,1].plot(x,y2,"b")
sfigs[0,1].set_title("second graph")

sfigs[1,0].plot(x,y3,"g")
sfigs[1,0].set_title("third graph")

sfigs[1,1].plot(x,y4,"orange")
sfigs[1,1].set_title("fourth graph")
```

# Scatter plot with displaying the categories in another variable

```python
cls = np.random.choice(a=["a", "c", "b"],size=100,p=[0.5, 0.3, 0.2])
height=np.random.normal(160,5,100)
weight=np.random.normal(55,2,100)

df=pd.DataFrame({"Height":height,"Weight":weight,"Class":cls})
df.head()
```

```python
col=["red" if i=="a" else "green" if i=="b" else "purple" for i in df["Class"]]
print(col)
```

```python
plt.scatter(df["Height"],df["Weight"],c=col)
plt.show()
```
